## 거리계산

```python
# Input: 리스트, 원소별로 곱셈하고 싶은 수
# Output: 각 원소에 num만큼 곱해진 리스트가 반환
def multiple(raw_list, num):
    new_list = []
    for i in raw_list:
        new_list.append(i * num)
    return new_list
```

```python
# 두 dataframe의 RGB값을 유클리디안 거리 계산을 이용하여 행렬 생성
# Input: 한 장르에 해당하는 모든 웹툰의 상위 6개 RGB값이 있는 DataFrame, 한 웹툰의 상위 6개 RGB값이 있는 DataFrame
# Output: min-max scailed Matrix(모든 원소가 0과 1사이 값)
def cal_matrix(df_genre, df_toon):
    matrix = []
    idx_list = []
    
    for i in range(len(df_genre)):
        idx = '{},{},{}'.format(df_genre.loc[i, 'R'], df_genre.loc[i, 'G'], df_genre.loc[i, 'B'])
        idx_list.append(idx)
        color_list = []
        m = []
        for j in range(len(df_toon)):
            color = '{},{},{}'.format(df_toon.loc[j, 'R'], df_toon.loc[j, 'G'], df_toon.loc[j, 'B'])
            color_list.append(color)
            distance = np.sqrt((df_genre.loc[i,'R']-df_toon.loc[j,'R'])**2 \
                            + (df_genre.loc[i,'G']-df_toon.loc[j,'G'])**2 \
                            + (df_genre.loc[i,'B']-df_toon.loc[j,'B'])**2)
            m.append(distance)
  
        # 각 웹툰의 별점을 가중치로 곱해줌. 
        m = multiple(m, df_genre.loc[i,'star_point'])
        matrix.append(m)
        
    matrix = pd.DataFrame(matrix)
    #print(matrix)
    
    #min-max scaling
    max_rgb = np.sqrt((255)**2 + (255)**2 + (255)**2)*10
    matrix = matrix / max_rgb
    return matrix
```

```python
# DataFrame의 모든 값을 더해서 평균냄.
# Input: cal_matrix를 통해 구한 행렬 DataFrame
# Output: 거리기반 점수(0과 1사이 값) -> 유사도와 반대개념이기 때문에 숫자가 작을수록 유사하다는 것을 의미.
def mean_df(df):
    sum_list = []
    for i in df:
        col = sum(df.loc[:,i])
        sum_list.append(col)
    sum_col = sum(sum_list)
    return sum_col / df.size
```

```python
# 모든 웹툰을 각 장르에 맞게 유사도 점수 계산을 함.
# Input: 한 장르에 해당하는 모든 웹툰의 상위 6개 RGB값이 있는 DataFrame, 점수를 구하고 싶은 웹툰, 빈 리스트
# Output: 웹툰명, 장르, 유사점수(0일수록 유사)가 들어있는 2차원 리스트 반환
def webtoon_score(df_genre, df_toon, score_list):
    for title in df_toon.title.unique():
        print(title)
        row = []
        row.append(title)
        toon = df_toon[df_toon.title==title].reset_index()
        matrix = cal_matrix(df_genre, toon)
        #print(matrix)
        score = mean_df(matrix)
        #print(score)
        row.append(score)
        for k in df_toon[df_toon.title==title].Genre:
            genre = k
        row.append(genre)
        #print(row)
        score_list.append(row)
    return score_list
```
